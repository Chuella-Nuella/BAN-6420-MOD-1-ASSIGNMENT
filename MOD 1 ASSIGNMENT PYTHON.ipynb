```python
#PYTHON CODE
import random
#Define constants
MIN_SALARY=7500
MAX_SALARY=30000
NUM_WORKERS=400
```

```python
# Define worker class
class Worker:
    def __init__(self, name, salary, gender):
        self.name = name
        self.salary = salary
        self.gender = gender
        self.employee_level = ""

    def calculate_employee_level(self):
        if self.salary > 10000 and self.salary < 20000:
            self.employee_level = "A1"
        elif self.salary > MIN_SALARY and self.salary < MAX_SALARY and self.gender == "Female":
            self.employee_level = "A5-F"
```

```python
# Generate workers dynamically

workers = []
for i in range(1, 401):
    name = f"Worker {i+1}"
    salary = random.uniform(MIN_SALARY, MAX_SALARY)
    gender = random.choice(["Male", "Female"])
    worker = Worker(name, salary, gender)
    workers.append(worker)

```

```python
# Generate payment slips
for worker in workers:
    try:
        worker.calculate_employee_level()
        print(f"Payment Slip for {worker.name}:")
        print(f"Salary: ${worker.salary:.2f}")
        print(f"Employee Level: {worker.employee_level}")
        print("-" * 30)
    except Exception as e:
        print(f"Error generating payment slip for {worker.name}: {str(e)}")
```
